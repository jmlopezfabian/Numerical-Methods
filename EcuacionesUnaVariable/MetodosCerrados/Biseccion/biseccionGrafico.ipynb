```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x**3 + x - 1
```

```python
def biseccion(f,a,b,tol):

    x = np.linspace(a, b, 1000)
    y = f(x)

    plt.plot(x, y)
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.axvline(x=a, linestyle='--', color='red')
    plt.axvline(x=b, linestyle='--', color='red')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Método de la bisección')
    
    if(f(a) * f(b) > 0):
        return None
    else:
        i = 0
        while ((b-a)/2) > tol:
            p = (a + b) / 2
            if(f(p) == 0):
                return p
            elif(f(a) * f(p) < 0):
                b = p
            else:
                a = p
            plt.plot(p, f(p), 'ro')
            i+=1

    plt.show()
    print(f"Raiz en: {p}")
    print(f"Iteraciones: {i}")
```

```python
biseccion(f,0,1,0.0005)
```
