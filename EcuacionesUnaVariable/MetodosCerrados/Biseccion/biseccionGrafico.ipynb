```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x**3 - x**2 - 100
```

```python
def biseccion(f,a,b,error,iter):

    x = np.linspace(a, b, 1000)
    y = f(x)

    plt.plot(x, y)
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.axvline(x=a, linestyle='--', color='red')
    plt.axvline(x=b, linestyle='--', color='red')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('Método de la bisección')
    
    if(f(a) * f(b) > 0):
        print(f"No hay raiz en el intervalo [{a},{b}]")
        return None
    else:
        #i = 0
        ##while ((b-a)/2) > tol:
        ##    p = (a + b) / 2
        ##    if(f(p) == 0):
        ##        return p
        ##    elif(f(a) * f(p) < 0):
        ##        b = p
        ##    else:
        ##        a = p
        ##    plt.plot(p, f(p), 'ro')
        ##    i+=1
        for i in range(iter):
            p = (b+a) / 2
            plt.plot(p, f(p),"ro")
            if abs((p - a) / p) < error or f(p) == 0:
                break
            elif f(a) * f(p) < 0:
                b = p
            else:
                a = p
    plt.show()
    print(f"Raiz en: {p}")
    print(f"Iteraciones: {i+1}")
    
    return p
    
```
